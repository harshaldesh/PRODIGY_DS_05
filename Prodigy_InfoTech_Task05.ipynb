```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
Acc=pd.read_csv('US_Accidents_March23.csv')
```

```python
Acc.isna().sum().sort_values(ascending=False).head(10)
```

```python
pd.set_option('display.max_columns', None)
```

```python
Acc.head()
```

```python
Acc= Acc.drop(columns=['End_Lat', 'End_Lng'])
```

```python
Acc.head(10)
```

```python
Acc['Start_Time'] = pd.to_datetime(Acc['Start_Time'], errors='coerce')
```

```python
Acc.head(15)
```

```python
Acc['Weather_Condition'] = Acc['Weather_Condition'].fillna('Unknown')
```

```python
Acc["Weather_Condition"]
```

```python
Acc.head()
```

```python
Acc.tail()
```

```python
Acc.head()
```

Minimal preprocessing was performed to ensure accurate visualizations. Non-essential columns were removed, and missing values were handled only where required for analysis.

```python
Acc['Hour'] = Acc['Start_Time'].dt.hour
```

```python
Acc['Hour']
```

```python
Acc.isna().sum().sort_values(ascending=False) * 100. / len(Acc)
```

```python
 #Accidents by Time of Day

Accidents_by_hour=Acc.groupby('Hour').size()

Accidents_by_hour.plot(kind='line',marker='o')
plt.title('Accidemts By Hour Of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
#Day Vs Night Accidents

Acc['Sunrise_Sunset'].value_counts().plot(kind='bar')
plt.title("Day vs Night Accidents")
plt.xlabel("Time of Day")
plt.ylabel("Accident Count")
plt.xticks(rotation=0)
plt.show()
```

```python
#Weather Condition Analysis

top_weather=Acc['Weather_Condition'].value_counts().head()

top_weather.plot(kind='bar')
plt.title("Top Weather Condition During Accidents")
plt.xlabel("Weather Condition")
plt.ylabel("Accidents Count")
plt.xticks(rotation=0)
plt.show()
```

```python
#Road Condition / Infrastructure Factors

road_features = {
    'Traffic Signal': Acc['Traffic_Signal'].sum(),
    'Junction': Acc['Junction'].sum(),
    'Crossing': Acc['Crossing'].sum(),
    'Roundabout': Acc['Roundabout'].sum()
}

pd.Series(road_features).plot(kind='bar')
plt.title("Accidents Near Road Features")
plt.ylabel("Accident Count")
plt.xticks(rotation=0)
plt.show()
```

```python
sample_Acc = Acc.sample(20000, random_state=42)

plt.scatter(sample_Acc['Start_Lng'], sample_Acc['Start_Lat'], s=1, alpha=0.3)
plt.title("Traffic Accident Hotspots")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
cities_by_accident = Acc.City.value_counts()
cities_by_accident[:20]
cities_by_accident[:20].plot(kind='barh')
```

```python
high_accident_cities = cities_by_accident[cities_by_accident >=1000] # having over 1000 accidents
low_accident_cities = cities_by_accident[cities_by_accident < 1000]  # having less than 1000 accidents
```

```python
# Percentage of high accident cities
len(high_accident_cities) / len(cities_by_accident)
```

```python
# Distribution of high accident cities
sns.distplot(high_accident_cities)
```

```python
# Distribution of high accident cities
sns.distplot(low_accident_cities)
```

```python
# Distribution of high accident cities
sns.histplot(high_accident_cities, log_scale=True)
```
